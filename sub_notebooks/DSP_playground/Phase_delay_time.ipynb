```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

```python
# create artifical signal
srate  = 1000 # hz
time   = np.arange(0.,2.,1/srate) # time vector in seconds
pnts   = len(time) # number of time points+
ph_shift = 41*np.pi/180
signal = 2.5 * np.sin( 2*np.pi*4*time ) # + 1.5 * np.sin( 2*np.pi*6.5*time )
signal2 = 2.5 * np.sin( 2*np.pi*4*time +ph_shift)# + 1.5 * np.sin( 2*np.pi*6.5*time +ph_shift)
```

```python
# general the delay should be
f = 4
dt = ph_shift/(2*np.pi*f)
dt
```

```python
plt.plot(time,signal)
plt.plot(time,signal2)
plt.xlim((0.5,0.75))
```

```python
p = sig.find_peaks(signal)[0]
p2 = sig.find_peaks(signal2)[0]
plt.plot(time,signal)
plt.plot(time,signal2)
plt.plot(time[p],signal[p],'+')
plt.plot(time[p2],signal2[p2],'+')
dt = time[p[0]]-time[p2[0]]
print(dt)
ph_from_dt = (dt*f*2*np.pi)
print(ph_from_dt*180/np.pi)
```
