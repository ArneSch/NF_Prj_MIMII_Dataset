```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create artifical signal
srate  = 100 # hz
time   = np.arange(0.,2.,1/srate) # time vector in seconds
pnts   = len(time) # number of time points+
ph_shift = 90*np.pi/180
signal = 2.5 * np.sin( 2*np.pi*4*time ) + 1.5 * np.sin( 2*np.pi*6.5*time )
signal2 = 2.5 * np.sin( 2*np.pi*4*time +ph_shift) + 1.5 * np.sin( 2*np.pi*6.5*time +ph_shift)
time_delay = 0.12
time_delay_sample=int(np.round(time_delay*srate))
signal3 = np.concatenate((np.zeros(time_delay_sample) ,signal[:-time_delay_sample]), axis=0)
time_delay_sample
```

```python
plt.plot(time,signal)
#plt.plot(time,signal2)
plt.plot(time,signal3)
```

```python
pad_tt = 2
pad_t = int(np.round(pad_tt*srate))
pad_sig1= np.concatenate((np.zeros(pad_t) ,signal, np.zeros(pad_t)), axis=0)
pad_sig2= np.concatenate((signal3, np.zeros(pad_t*2)), axis=0)
#pad_time = np.concatenate((np.linspace(-pad_tt,0,pad_t) ,time, np.linspace(time,0,pad_t)
plt.plot(pad_sig1)
#plt.plot(time,signal2)
plt.plot(pad_sig2)
```

```python
tv = np.array([])
cv = np.array([])
for t in np.linspace(0.01,5,512):
    td = int(np.round(t*srate))
    s2 = np.concatenate((np.zeros(td) ,pad_sig2[:-td]), axis=0)
    c = np.correlate(pad_sig1,s2)
    tv = np.append(tv,[t])
    cv = np.append(cv,[c])
    #print (t,c)
 
tv.shape
plt.plot(tv,cv)
np.mean(tv[np.where(cv == np.max(cv))]*-1+pad_tt)
```
