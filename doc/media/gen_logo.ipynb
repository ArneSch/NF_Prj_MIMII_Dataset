```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Surface(z=mel_data,
                  colorscale='Inferno')])

fig.update_layout(title='mel spectrum', 
                  autosize=False,
                  width=1200, 
                  height=1200,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

```python
import plotly.graph_objects as go
import plotly.express as px

z = [mel_data[i] for i in range(mel_data.shape[0])]
y = [range(mel_data.shape[1]) for i in range(mel_data.shape[0])]
x = [[i]*mel_data.shape[1] for i in range(mel_data.shape[0])]

fig = go.Figure()
for i in range(mel_data.shape[0]):
    fig.add_trace(px.line_3d(z=z[i],
                       y=y[i],
                       x=x[i]))
fig.show()
```
