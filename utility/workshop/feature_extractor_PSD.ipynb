# Concat channel

```python
BASE_FOLDER = '../../'
%run -i ..\feature_extractor\JupyterLoad_feature_extractor.py
file_path = r'\dataset\6dB\pump\id_00\normal\00000004.wav'
## test:
fe_psd_f = feature_extractor_welchPSD(BASE_FOLDER,'f')
fe_psd = feature_extractor_welchPSD(BASE_FOLDER,'f')
fe_psd_f.set_hyperparamter(channel='all',multichannel='concat')
fe_psd_f.create_from_wav(file_path)
fe_psd.create_from_wav(file_path)
fe_psd_f.plot()
print(fe_psd_f.feature_data['A'].shape)
fe_psd.plot()
print(fe_psd.feature_data['A'].shape)
```
